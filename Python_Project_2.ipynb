```python
# Import Pandas and Matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data set for NYC Public School Regents for the school years 2014-15 to 2016-17
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv', index_col=0)
```

```python
print("Column Names:")
print(df.columns)
```

```python
# Display the top 5 records
df.head()
```

```python
#Filtering data for Patients with Diabetes Medication and Removing Unknown/Invalid Gender
filtered_df = df[(df['diabetesMed'] == 'Yes') & (~df['gender'].isin(['Unknown/Invalid']))]
```

```python
#Analyzing data for Patients with Diabetes Medication by Race and Gender
Summary = filtered_df.pivot_table(
    index='race',
    columns='gender',
    aggfunc='size',
    fill_value=0
).reset_index()
print(Summary)
```

```python
#Setting race as index for plotting
Summary.set_index('race', inplace=True)
```

```python
#Creating a grouped bar chart
Summary.plot(kind='bar', figsize=(10, 6))
```

```python
#Drafting conclusion of my findings
print("Conclusion:\n\nBased on the dataset more Female than Male are taking Diabetes Medication.\n")
print("In addition, the race with the most individuals with Diabetes Medication is Caucasion followed by AfricanAmerican.\nAsian had the least individuals with Diabetes Medication")
```

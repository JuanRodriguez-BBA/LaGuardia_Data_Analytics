```python
# Creating a list for Odd and Even numbers
Odd_numbers = list(range(1,100001,2))
Even_numbers = list(range(2,100001,2))
```

```python
# Enter a number to check if Odd or Even number and a multiple of 4
Number = 17
if Number in Even_numbers and Number/ 4 == round(Number/4,0):
    print("This number is an even number and a multiple of 4")
elif Number in Even_numbers:
    print("This number is an even number")
else:
    print("This number is an odd number")
```

```python
# Please enter a number (check) to divide by (num)
check = 4
num = 2

if check / num == round(check / num,0):
    print(str(check) + " divides evenly into " + str(num) + "; the total is " + str(check/num))
else:
    print(str(check) + " does not divide evenly into " + str(num) + "; the total is " + str(round(check/num,2)))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
plt.style.use('ggplot')
pd.set_option('display.max_columns', 200)
import plotly.express as px
```

```python
#Loading Data from the International Organization for Migration for 2022 thru 2023
df1 = pd.read_csv('/Users/billonario/Downloads/MissingMigrants-Global-(2023-2022).csv')
```

```python
#Loading Data from the International Organization for Migration for 2024 thru 2025
df2 = pd.read_csv('/Users/billonario/Downloads/MissingMigrants-Global-2025-2024.csv')
```

```python
#Reviewing the total number of Columns and Rows for df2
df2.shape
```

```python
#Combining both df1 and df2 verticaly into a single dataframe called df
df = pd.concat([df1, df2])
```

```python
#Reviewing the total number of Columns and Rows for df
df.shape
```

```python
#Reviewing the first couple of rows
df.head(5)
```

```python
#Renaming fields with spaces to include "_"
df = df.rename(columns={'Main ID' : 'Main_ID','Incident ID' : 'Incident_ID',
'Region' : 'Region',
'Incident Date' : 'Incident_Date',
'Reported Month' : 'Reported_Month',
'Number Dead' : 'Number_Dead',
'Minimum Estimated Number of Missing' : 'Minimum_Estimated_Number_of_Missing',
'Number of Survivors' : 'Number_of_Survivors',
'Number of Females' : 'Number_of_Females',
'Number of Males' : 'Number_of_Males',
'Number of Children' : 'Number_of_Children',
'Region of Origin' : 'Region_of_Origin',
'Cause of Death' : 'Cause_of_Death',
'Country of Origin' : 'Country_of_Origin',
'Location of death' : 'Location_of_death',
'Information Source' : 'Information_Source',
'Coordinates' : 'Coordinates',
'Migration route' : 'Migration_route',
'UNSD Geographical Grouping' : 'UNSD_Geographical_Grouping',
'Source Quality' : 'Source_Quality'})
```

```python
df.head(20)
```

```python
#Identify columns with missing values
df.isna().sum()
```

```python
#Reviewing data for duplicates
df.loc[df.duplicated()]
```

```python
df.query('Migration_route == "Dominican Republic to Puerto Rico"')
```

```python
df.groupby(['Migration_route'])[['Number_Dead','Minimum_Estimated_Number_of_Missing','Number_of_Survivors']].sum().plot(kind='bar',figsize=(10,8))
```

```python
df.groupby(['Migration_route'])['Number_Dead'].sum().plot(kind='barh', figsize=(10,5), title='Missing Migrants by Migration Route',xlabel='Number of reported deaths',ylabel='Migration Rout')
```

```python
df[['latitude','longitude']] = df['Coordinates'].str.split(',', expand=True)
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')
df['Cause_of_Death'] = pd.to_numeric(df['Cause_of_Death'], errors='coerce')
```

```python
df.columns.tolist()
```

```python
px.scatter_mapbox(df, lat='latitude',
                  lon='longitude',
                  mapbox_style='open-street-map',
                  width = 1200,
                  height = 600,
                  zoom = 2.5,
                  color = 'Cause_of_Death',
                  size = 'Number_Dead',
                  size_max = 50,
                  title='Missing Migrants 2022 thru 2025'
                 )
```
